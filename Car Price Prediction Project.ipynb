## Model

```python
X=car.drop(columns="Price")
y=car["Price"]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
X_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder #linear regression requires numerical columns
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder(handle_unknown='error')
ohe.fit(X[["name","company","fuel_type"]]) #use fit-transform when u r not using any algo
```

```python
ohe.categories_
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),["name","company","fuel_type"]),remainder="passthrough")
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr) 
#one end of pipeline will go to column transformer and other end will get predictions from linear regression model
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
y_pred
```

```python
X_test
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr) 
    pipe.fit(X_train, y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
    
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr) 
pipe.fit(X_train, y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
